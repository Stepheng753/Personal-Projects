### Math

Notice the following: 
> A single pixel can be represented as an integer value ranging from 0 to 255 inclusive
<br>
> An image can be represented as 2D Matrix that consists of $n \times m$ pixels.

Let the following be true:
<br>
> $X$ will represent a 2D Matrix in which each column will represent an image in a 1D form 
<br>
> Each image $x^{(k)}$, where 1 < k < m, is an array of $16 \times 16 = 784$ values
$$
X = 
\begin{bmatrix}
\cdots & x^{(1)} & \cdots \\
\cdots & x^{(2)} & \cdots \\
\cdots & \vdots & \cdots \\
\cdots & x^{(m - 1)} & \cdots \\
\cdots & x^{(m)} & \cdots
\end{bmatrix}^T = 
\begin{bmatrix}
\vdots & \vdots & \vdots & \vdots & \vdots \\
x^{(1)} & x^{(2)} & \cdots & x^{(m - 1)} & x^{(m)}\\
\vdots & \vdots & \vdots & \vdots & \vdots
\end{bmatrix}
$$
